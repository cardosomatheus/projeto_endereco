```python
import pandas as pd

#   ABERTURA DO ARQUIVO
dados = pd.read_csv('tbl_cep_201805_csv.csv',encoding="ISO-8859-1")
```

```python
#   FILTRANDO AS TABELAS CAMPO NÃO NULOS
dados_filtrado = dados[['cep','logradouro','bairro','cidade', 'estado','latitude','longitude','cidade_ibge','cidade_area']]
dados_nao_nulos = dados_filtrado[~dados_filtrado['bairro'].isnull()]
bairro_nao_nulo = dados_nao_nulos[~dados_nao_nulos['logradouro'].isnull()]
```

```python
bairro_nao_nulo.drop_duplicates(subset='cep')
```

```python
cidades = bairro_nao_nulo[['cidade','estado','cidade_ibge','cidade_area']]
cidades=cidades.drop_duplicates(subset='cidade')
```

```python
cidades.to_csv('cidades.csv',sep=';', header=False, index=False)
```

```python
#   exportação em csv
bairro_nao_nulo.to_csv('informacoes_bairros', sep=';', header=False, index=False)
```
