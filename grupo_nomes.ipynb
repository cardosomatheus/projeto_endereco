```python
import pandas as pd
from conexao_oracle import conexao_local_host
import random as rn 
from datetime import date
import sqlalchemy as sa
#Leitura do csv
dados = pd.read_csv('grupos.csv')
```

```python
# GERA UM DATA DE NASCIMENTO COM IDADE MAXIMO DE 20 E MINIMA DE 18
def informa_dt_nascimento(min_year = (date.today().year - 80), max_year = (date.today().year - 18) ):
    # gera um datetime no formato yyyy-mm-dd hh:mm:ss.000000
    ano = rn.randint(min_year, max_year)
    mes = rn.randint(11, 12)
    dia = rn.randint(1, 28)
    data = f'{ano}-{mes}-{dia}'
    return data

```

```python
#   definindo as tabelas que serão ultilizadas
dados = dados[['name','classification']]
dados = dados.dropna(subset=['name','classification'])
dados['data_nascimento'] = 0

# ALTERANDO O TIPO DE CADA SERIES
dados['data_nascimento'] = pd.to_datetime(dados['data_nascimento'].apply(lambda x: informa_dt_nascimento()))
dados['name'] = dados['name'].astype(dtype='string')
dados['classification'] = dados['classification'].astype(dtype='string')


```
